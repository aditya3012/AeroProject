```python
#transforming into Joukowsky 
 
N_panels = 50
T  = 0.12
c = 1 
H = 0.02
a = c/4
theta = numpy.linspace(0, 2*math.pi, N_panels+1) 
Zeta_c = -4*a/(3*math.sqrt(3))*T+1j*2*H
r_s = ((1/4)+(T/(3*math.sqrt(3))))*c
 
Zeta = (r_s)*numpy.exp(1j*(theta))

J = (Zeta+Zeta_c) + a**2/(Zeta+Zeta_c)

x_ends = J.real
y_ends = J.imag

#x_ends = x_ends[::-1]
#y_ends = y_ends[::-1]


```

```python
N_panels = 50
T  = 0.12
c = 0.3 
H = 0.0
a = c/4
theta = numpy.linspace(0, 2*math.pi, N_panels+1) 
Zeta_c = -4*a/(3*math.sqrt(3))*T+1j*2*H
r_s = ((1/4)+(T/(3*math.sqrt(3))))*c
 
Zeta = (r_s)*numpy.exp(1j*(theta))

J1 = (Zeta+Zeta_c) + a**2/(Zeta+Zeta_c)

x1_ends = J1.real+0.6
y1_ends = J1.imag-0.05
```

```python
# #transforming into Joukowsky 
pyplot.figure(figsize=(17,5))

pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)

pyplot.plot(x_ends, y_ends, color='#CD2305', linestyle='-', linewidth=2)
pyplot.plot(x1_ends, y1_ends, color='#CD2305', linestyle='-', linewidth=2)
pyplot.axis("equal")
```
