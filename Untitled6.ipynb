# 1. For a Potential Flow vortex, Velocity varies with "r" 

A potential vortex is the flow in circular paths around the circular path.

considering in a cylidrical coordinate system, in 2D
 the velocity vector V= (Ur, U0, Uz)
 
   therefore V = urer + utet + uzez
             
         but Urer = 0 (since no radial velocity)
             uzez = 0 ( it is 2-d)
             
      therefore V = U0
      
       U0 = velocity of the vortex.
       
       derivation for U0
       
       since the flow is irrotational all components of the vorticity vector are zero.
       
       therefore, the vorticity in coordinates is 
       
                   ∇×v = (1/r* ∂uz/∂θ - ∂uθ/∂z) Xr + (∂ur/∂z- ∂uz/∂r)Xθ +(1/r* ∂ruθ/∂r - 1/r*∂ur/∂uθ ) Xz 
                   
                   Ur^2 = Ux^2 + uy^2
                   
                   Uθ = UxCos θ
                   
              since the vortex is 2D the z component becomes "0" and 
              and its a irrotational flow.
              
              therefore, ∂ruθ/∂r - ∂ur/∂uθ = 0
              
                         ∂ruθ/∂r = ∂ur/∂uθ = 0
                         
                     on solving we get,
                     
                         uθ = k/r
                         
               therefore the velocity varies with radius.

```python
def get_velocity_vortex (strength, xv, yv, X, Y):
     
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v
```

```python
# Doublet Strength and location
K = 1.0
x_doublet, y_doublet = 0.0, 0.0
```

```python
#Velocity of the doublet

def get_velocity_doublet(strength, xd, yd, X, Y):

  u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
  v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
  return u, v
```

```python
# Stream Function of the doublet

def get_stream_function_doublet(strength, xd, yd, X, Y):
    
    psi= -strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    
    return psi
```

```python
# computes the velocity field on the mesh grid
u_doublet, v_doublet = get_velocity_doublet(K, x_doublet, y_doublet, X, Y)

# computes the stream-function on the mesh grid
psi_doublet = get_stream_function_doublet(K, x_doublet, y_doublet, X, Y)
```
