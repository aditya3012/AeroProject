```python
import math
import numpy
from matplotlib import pyplot
```

```python
N = 50                            
x_start, x_end = -6.0, 6.0            
y_start, y_end = -4.0, 4.0            
x = numpy.linspace (x_start, x_end, N)    
y = numpy.linspace (y_start, y_end, N)    
X, Y = numpy.meshgrid(x, y)

print('x = ', x)
print('y = ', y)

X, Y = numpy.meshgrid(x, y)
```

```python
%matplotlib inline


size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.scatter(X, Y, s=10, color='#CD2305', marker='o', linewidth=0)
```

```python
#when the strength of the source and sink are equal
```

```python
# To plot the stream lines for the source  

strength_source = 2
x_source, y_source = -2 , 0

u_source = strength_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = strength_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
```

```python
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_source, v_source, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
#To plot the streamlines of the sink 

strength_sink = -2.0                     
x_sink, y_sink = 2.0, 0.0                


u_sink = strength_sink/(2*math.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = strength_sink/(2*math.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)

```

```python
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_sink, v_sink, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_sink, y_sink, color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
#computes the velocity of the pair by superposition

u_pair = u_source + u_sink
v_pair = v_source + v_sink

#plots the stram lines of the pair 

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_pair, v_pair, density=2.0, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter([x_source, x_sink], [y_source, y_sink], 
            color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
# when the strength of the sources and sink are equal,the line between the source and sink indicates the deviding streamline and 
# it forms rankine oval. All the stream lines inside the oval originate at the left and flow into the sink on the right the net 
# outflow from the oval is zero
```

```python
#plots the contours of the velocity potential
size = 10
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, u_pair, v_pair, levels=numpy.linspace(-2.0, 2.0, 10), extend='both')

pyplot.scatter([x_source, x_sink], [y_source, y_sink], color='#CD2305', s=120, marker='o')
pyplot.contour(X, Y, u_pair, v_pair, levels=[0.2], colors='#CD2305', linewidths=2, linestyles='solid');
```
