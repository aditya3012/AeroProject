```python
import numpy as npy
from matplotlib import pyplot as plt
from math import *
```

```python
N = 50

c = 1.0
T = 0.05
H = 0.00
a = c/4

r_s=sqrt(((c+.1)**2 + .1**2))

r_start, r_stop = r_s,2*r_s
theta_start, theta_end = 0, 2*pi           

R = npy.linspace (r_start, r_stop, N)    
theta = npy.linspace (theta_start, theta_end, N)    

R, theta = npy.meshgrid(R, theta)

x=np.cos(theta)*R + c-r_s
y=np.sin(theta)*R

z = npy.zeros_like(x, dtype = complex)

z.real=x
z.imag=y

%matplotlib inline
plt.figure()
plt.scatter(z.real,z.imag)
plt.show()
```

```python
z_2= z  + (c**2)/z

%matplotlib inline
plt.figure()
plt.scatter(z_2.real, z_2.imag)
plt.scatter(z_2.real[:,0],z_2.imag[:,0], color='r')
plt.show()
```
