# 1. a.Formula for the radius of the circle generated by doublet.
               
            Let the strength of the Source be σ located at (-l/2,0)
                and sink strength be -σ located at (l/2,0).
                
            The Stream function evaluated at a point
                
                ψ(x,y) = σ / 2π (θ1−θ2)
                
                       = −σ / 2πΔθ 
                       
             When the seperation between the two points approach zero,
             simultaneously the strength of the source and sink are increased.
             so the product σl remains constant
              
                      K = σl 
                      
              The resultant flow is the doublet with strength K.
              
                  ψ(x,y)= lim (- σ / 2π dθ) and  σl= const 
                          l→0
                          
                          when dθ is infinitesimal,
                          
                          a = l sinθ
                          b = r-lcosθ
                         dθ = a/b 
                         
              Therefore the stream function
                          
                          ψ(r,θ) = lim (−σlsinθ / 2πr−lcosθ) 
                                   l→0
                          ψ(r,θ) = −κ sinθ / 2πr
                          
               when the uniform flow is superimposed with a doublet,
               
                          ψ = {V∞ r sin θ − (κ sin θ / 2π r)}
                          
                            = V∞ r sin θ { 1 - κ / 2π V∞ r^2)}
                            
                          ψ =  V∞ r sin θ {1 − (R^2 / r^2)}
                          
                          radius R^2 = k / (2πV∞)

            

```python
import math
import numpy
from matplotlib import pyplot
```

```python
# Creating Meshgrid
N = 50
x_start, x_end = -2.0,2.0
y_start, y_end = -1.0,1.0
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x,y)
```

```python
# Doublet Strength and location
K = 1.0
x_doublet, y_doublet = 0.0, 0.0
```

```python
#Velocity of the doublet

def get_velocity_doublet(strength, xd, yd, X, Y):

  u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
  v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
  return u, v
```

```python
# Stream Function of the doublet

def get_stream_function_doublet(strength, xd, yd, X, Y):
    
    psi= -strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    
    return psi
```

```python
# computes the velocity field on the mesh grid
u_doublet, v_doublet = get_velocity_doublet(K, x_doublet, y_doublet, X, Y)

# computes the stream-function on the mesh grid
psi_doublet = get_stream_function_doublet(K, x_doublet, y_doublet, X, Y)
```

```python
%matplotlib inline

# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_doublet, v_doublet, 
               density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o');
```

## uniform flow over Doublet

```python
#velocity of a uniform flow is 1

u_inf = 1.0
u_freestream = u_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)

psi_freestream = u_inf * Y
```

```python
# superposition of the doublet on the freestream flow
u = u_freestream + u_doublet
v = v_freestream + v_doublet
psi = psi_freestream + psi_doublet

# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')

# calculates the stagnation points
x_stagn1, y_stagn1 = +math.sqrt(K/(2*math.pi*u_inf)), 0
x_stagn2, y_stagn2 = -math.sqrt(K/(2*math.pi*u_inf)), 0

# adds the stagnation points to the figure
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```

#  2. a. Coeeficient of Pressure on the surface of a circle

                Let the surface of the radius of the cylinder be 'R'
                
                The redial velocity       Vr = 0 -------------> 1
                
                The Tangential velocity   Vt = −2V∞ sin θ ----> 2
                
                at θ= +/-90 we have the maximum speed
                
                therefore, Vt = 2V∞
                
                on applying the bernoulli equation;
                
                    p(θ) = p' − 1/2 ρ ( Vr^2 + Vt^2) --------> 3
                    
                     since, Vr = 0 and Vt = -2V∞ sin θ
                     
                     plugging the values in equation 3
                     
                     we get, p' = P∞ + 1/2 ρ V∞^2 (1-4sin^2 θ )
                            
                             (P'- P∞ / 1/2 ρ V∞^2 ) = 1-4sin^2 θ
                             
                             (P'- P∞ / 1/2 ρ V∞^2 ) = Cp
                             
                             therefore Cp = 1- 4sin^2 θ
                     




```python
def stream_function_doublet_cylnder(strength, xd, yd, X, Y):
  
    r = np.sqrt((X - xd)**2 + (Y - yd)**2)
    theta = np.arctan(Y/X)
    
    psi = - strength/(2*math.pi)*((np.sin(theta))/r)
    
    return psi
```

```python
def stream_function_uniform_flow_cylnder(u_inf, xd, yd, alpha, X, Y):
    
    r = np.sqrt((X - xd)**2 + (Y - yd)**2)
    theta = np.arctan(Y/X)
    
    psi = u_inf * (((r*np.sin(theta))*np.cos(alpha))\
                   - ((r*np.cos(theta))*np.sin(alpha)))
    
    return psi
```

```python
def vel_uniform_flow(u_inf, xd, yd, alpha, N, X, Y):

    r = np.sqrt((X - xd)**2 + (Y - yd)**2)
    theta = np.arctan(Y/X)
    
    # computes the freestream velocity field
    v_r = u_inf * np.cos(theta - alpha)
    v_theta = - u_inf * np.sin(theta - alpha)
    
    return v_r, v_theta
```

```python
def vel_doublet_cylinder(strength, xd, yd, X, Y):

    r = np.sqrt((X - xd)**2 + (Y - yd)**2)
    theta = np.arctan(Y/X)
    
    v_r = - strength/(2*math.pi)*((np.cos(theta))/r**2)
    v_theta = - strength/(2*math.pi)*((np.sin(theta))/r**2)
    
    return v_r, v_theta
```

```python
#calling doublet streamfunctions in cylindrical
psi_doublet_cylinder = stream_function_doublet_cylnder(K, x_doublet, y_doublet, X, Y)

#calling uniform flow streamfunctions in cylindrical
psi_uniform_flow_cylinder = stream_function_uniform_flow_cylnder(u_inf, x_doublet, y_doublet, alpha, X, Y)

#calling doublet velocities in cylindrical
v_r_doub, v_theta_doub = vel_doublet_cylinder(K, x_doublet, y_doublet, X, Y)

#calling uniform flow velocities in cylindrical
v_r_uni, v_theta_uni = vel_uniform_flow(u_inf, x_doublet, y_doublet, alpha, N, X, Y)
```

```python
#doublet and freestream combination
v_r = v_r_doub + v_r_uni
v_theta = v_theta_doub + v_theta_uni
psi = psi_uniform_flow_cylinder + psi_doublet_cylinder

# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, v_r, v_theta, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')

# calculates the stagnation points
x_stagn1, y_stagn1 = +math.sqrt(K/(2*math.pi*u_inf)), 0
x_stagn2, y_stagn2 = -math.sqrt(K/(2*math.pi*u_inf)), 0

# adds the stagnation points to the figure
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```

## Coefficient of pressure 

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

# plots the pressure coefficient field
size = 10
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp, levels=numpy.linspace(-2.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')
pyplot.contour(X,Y,psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```
