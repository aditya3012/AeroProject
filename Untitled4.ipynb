# 1.  Consider the flow field as shown in figure, Given Source of a strength at location σ = 2 (− 2,0), sink of strength σ = −2 at location (2,0) 

```python
import math 
import numpy
from matplotlib import pyplot
```

```python
N = 100                                # number of points in each direction
x_start, x_end = -4, 4.0            # boundaries in the x-direction
y_start, y_end = -4.0, 4.0            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates

print('x = ', x)
print('y = ', y)

X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
%matplotlib inline

# plots the grid of points
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.scatter(X, Y, s=10, color='#CD2305', marker='o', linewidth=0)
```

## For the source flow

```python

strength_source = 2.0                      # source strength
x_source, y_source = -2.0, 0.0             # location of the source

# computes the velocity field on the mesh grid
u_source = strength_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = strength_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
```

## let's plot the streamlines from the source.

```python
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_source, v_source, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
```

## Sink Flow

```python
strength_sink = -2.0                     # strength of the sink
x_sink, y_sink = 2.0, 0.0                # location of the sink

# computes the velocity on the mesh grid
u_sink = strength_sink/(2*math.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = strength_sink/(2*math.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
```

```python
# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_sink, v_sink, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_sink, y_sink, color='g', s=80, marker='o', linewidth=0);
```

## source and sink pair

```python
# computes the velocity of the pair source/sink by superposition
u_pair = u_source + u_sink
v_pair = v_source + v_sink

# plots the streamlines of the pair source/sink
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_pair, v_pair, density=2.0, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter([x_source, x_sink], [y_source, y_sink], 
            color= ['#CD2305', 'g'],  s=80, marker='o', linewidth=0);
```

# To show the dividing streamlines for a source are given by Ψ = ±σ /2 .

```python
N = 50                                # number of points in each direction
x_start, x_end = -4.0, 4.0            # boundaries in the x-direction
y_start, y_end = -2.0, 2.0            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates

print('x = ', x)
print('y = ', y)

X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
%matplotlib inline

# plots the grid of points
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.scatter(X, Y, s=10, color='#CD2305', marker='o', linewidth=0)
```

```python
u_inf = 1.0        # freestream speed

# computes the freestream velocity field
u_freestream = u_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)

# computes the stream-function
psi_freestream = u_inf * Y
```

```python
def get_velocity(strength, xs, ys, X, Y):
   
    u = strength/(2*numpy.pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v = strength/(2*numpy.pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    
    return u, v
```

```python
def get_stream_function(strength, xs, ys, X, Y):
   
    psi = strength/(2*numpy.pi)*numpy.arctan2((Y-ys), (X-xs))
    
    return psi
```

```python

strength_source = 5.0                      # source strength
x_source, y_source = -1.0, 0.0             # location of the source

# computes the velocity field on the mesh grid
u_source = strength_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = strength_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
```

```python
# superposition of the source on the freestream
u = u_freestream + u_source
v = v_freestream + v_source
psi = psi_freestream + psi_source

# plots the streamlines
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o')

# calculates the stagnation point
x_stagnation = x_source - strength_source/(2*numpy.pi*u_inf)
y_stagnation = y_source

# adding the stagnation point to the figure


# adds the dividing line to the figure
pyplot.contour(X, Y, psi, 
            levels=[-strength_source/2, +strength_source/2], 
            colors='#CD2305', linewidths=2, linestyles='solid');
```

From the superposition principle we have: 
                            ψ= ψfreestream + ψ source
                            ψfreestream= U(y-ysource ) 
                            ψsource = σ/2 π arctan (y-ysource/x-xsource)
                            
                            considering in polar coordinates 
                            y-ysource = rsinθ
                            X-xsource = rcosθ
                            
                            therefore plugging in the equation 1
                            
                           ψ = Ursinθ+σ/2 π arctan (rsinθ/rcosθ)
                               Ursinθ+σ/2 π arctan (tanθ)
                               
                               Ursinθ= 0 (since there is no velocity at the deviding streamline.)
                               
                         therefore  ψ = σ/2 π θ (but θ = π)  
                                    ψ = σ/2 π (hence proved)

# 3. To Derive The formulas for length and width of a Rankine Circle

## Derivation: 
                  Flow generated by the source (σ ),located at a point (x,y) be (-a,0) and at sink (-σ), located at a point (X,Y) be (a,0) and Uniform stream U. This creates a closed streamline aroud a source and sink pair, called Rankine circle.
                  
## resulting Stream Function: ψ = Uy − m tan−1 (2ay/x2 + y2 − a2 )  -----------> 1
                  
                  Flow seperates after meeting the inflow from the source, creating a stagnation point at "X=-D" and similarly it goes around the source and coverges due to the influence of the sink. Therefore, it creates a stagnation point at "X=D". When these two stagnation points are coneected by the streamline they form the closed body called as Rankine Oval.
                  
## To determine the streamlines where the stagnation point occurs:

### since both the points are at Y=0; for which ψ = 0 in equation 1

###                                                therefore,        0 = Uy − σ  tan−1 (2ay/x2 + y2 − a2 ) ---------------> 2


## To determine the length of the Rankine oval;

###   Length 'l' of a Rankine circle can be determined by taking y<< 1

### therefore, equation 2 becomes 

###                                               Uy ≈ σ  tan−1 (2ay/D^2 − a^2 )

###                                               Uy ≈ σ  (2ay/D^2 − a^2) 

###                                                D ≈ [a^2 +2aσ /U]^1/2

###                                                D/a ≈ [1 +2σ /Ua]^1/2

## To determine the width 'h' of the rankine circle.

###                                               when x = 0

###                                               tan (Uh/σ) ≈ 2ah / h^2 − a^2 

###                                               tan [(Ua/σ)(h/a)] ≈ 2 h/a[ 1 / {(h/a)^2 − 1}]

###                                               h / a ≈ 1 / 2 [( h / a ) ^ 2 − 1 ] tan [(ua/σ)(h/a)]

#### Note; Due to Transcendental character of equation, height has to be calculate numerically.

# 4. Plot the Coefficient of Pressure Cp for the figure. plot filled contour plot and color bar.

```python
N = 50                                # number of points in each direction
x_start, x_end = -4.0, 4.0            # boundaries in the x-direction
y_start, y_end = -2.0, 2.0            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates

print('x = ', x)
print('y = ', y)

X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
%matplotlib inline

# plots the grid of points
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.scatter(X, Y, s=10, color='#CD2305', marker='o', linewidth=0)
```

```python
u_inf = 1.0        # freestream speed

# computes the freestream velocity field
u_freestream = u_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)

# computes the stream-function
psi_freestream = u_inf * Y
```

```python
def get_velocity(strength, xs, ys, X, Y):
    """Returns the velocity field generated by a source/sink.
    
    Arguments
    ---------
    strength -- strength of the source.
    xs, ys -- coordinates of the source.
    X, Y -- mesh grid.
    """
    u = strength/(2*numpy.pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v = strength/(2*numpy.pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    
    return u, v
```

```python
def get_stream_function(strength, xs, ys, X, Y):
    """Returns the stream-function generated by a source.
    
    Arguments
    ---------
    strength -- strength of the source.
    xs, ys -- coordinates of the source.
    X, Y -- mesh grid.
    """
    psi = strength/(2*numpy.pi)*numpy.arctan2((Y-ys), (X-xs))
    
    return psi
```

```python
strength_source = 5.0            # strength of the source
x_source, y_source = -1.0, 0.0   # location of the source

# computes the velocity field
u_source, v_source = get_velocity(strength_source, x_source, y_source, X, Y)

# computes the stream-function
psi_source = get_stream_function(strength_source, x_source, y_source, X, Y)
```

```python
# superposition of the source on the freestream
u = u_freestream + u_source
v = v_freestream + v_source
psi = psi_freestream + psi_source

# plots the streamlines
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o')

# calculates the stagnation point
x_stagnation = x_source - strength_source/(2*numpy.pi*u_inf)
y_stagnation = y_source

# adding the stagnation point to the figure
pyplot.scatter(x_stagnation, y_stagnation, color='g', s=80, marker='o')

# adds the dividing line to the figure
pyplot.contour(X, Y, psi, 
            levels=[-strength_source/2, +strength_source/2], 
            colors='#CD2305', linewidths=2, linestyles='solid');
```

## To Plot a contour and Pressure bar

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

# plots the pressure coefficient field
size = 10
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp, levels=numpy.linspace(-2.0, 1.0, 50), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=20)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
pyplot.scatter([x_source, x_source], [y_source, y_source], color='#CD2305', s=80, marker='o')
pyplot.contour(X, Y, psi, levels=[-strength_source/2, +strength_source/2], colors='#CD2305', linewidths=1, linestyles='solid');
```
