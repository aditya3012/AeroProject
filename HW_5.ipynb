# 1. To calculate the circulation around the vortex

```python
N=100
x_start, x_end = -2.0,2.0
y_start, y_end = -1.0,1.0
x = numpy.linspace(x_start, x_end, N)    # computes a 1D-array for x
y = numpy.linspace(y_start, y_end, N)    # computes a 1D-array for y
X, Y = numpy.meshgrid(x, y)
```

```python
gamma = 5.0
x_vortex, y_vortex =0.0,0.0 
```

```python
def get_velocity_vortex (strength, xv, yv, X, Y):
     
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v
```

```python
def get_stream_function_vortex(strength, xv, yv, X, Y):
    
    psi = strength/(4*math.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    
    return psi
```

```python
# computes the velocity field on the mesh grid
u_vortex, v_vortex = get_velocity_vortex(gamma, x_vortex, y_vortex, X, Y)

# computes the stream-function on the mesh grid
psi_vortex = get_stream_function_vortex(gamma, x_vortex, y_vortex, X, Y)

```

```python
# plots the streamlines
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)




pyplot.streamplot(X, Y, u_vortex, v_vortex, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_vortex, y_vortex, color='#CD2305', s=80, marker='o');
```

# 3. Infinite row of vorticies

## a). To plot the flow field around finite number of vortices 

```python
import math 
import numpy
from matplotlib import pyplot as plt
```

```python
N = 100 #number of pts each direction
xs, xe = -2.0, 2.0 #x start and end
ys, ye = -0.5, 0.5 #y start and end
n = 3 #number of vortices
alpha = 0.0
u_inf = 1.0

xv = np.linspace(xs,xe,n)
yv = np.zeros_like(xv)
gamma = 5.0
vort_strength = gamma * np.ones_like(xv)
```

```python
def gen_uform_grid(N, x_s, x_e, y_s, y_e):
   
    x = np.linspace(x_s,x_e,N) #1-D array for x
    y = np.linspace(y_s,y_e,N) #1-D array for y
    X, Y = np.meshgrid(x,y)    #generates mesh grid
    
    return X, Y
```

```python
def vel_uniform_flow_alpha(u_inf, alpha, N, X, Y):
    
    # computes the freestream velocity field
    u = u_inf * (np.cos((alpha + np.zeros((N, N), dtype=float))))  #d(psi)/dy
    v = u_inf * (np.sin((alpha + np.zeros((N, N), dtype=float)))) #-d(psi)/dx
    return u, v
```

```python
def sf_uniform_flow_alpha(u_inf, alpha, X, Y):
    

    # computes the stream-function
    psi = u_inf * ((Y*np.cos(alpha)) - (X*np.sin(alpha)))

    return psi
```

```python

u_free, v_free = vel_uniform_flow_alpha(u_inf, alpha, N, X, Y)
psi_free = sf_uniform_flow_alpha(u_inf, alpha, X, Y)
```

```python
def vel_vortex(strength, xv, yv, X, Y):
    

    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v
```

```python
def sf_vortex(strength, xv, yv, X, Y):
    
    psi = strength/(2.0*math.pi)*np.log(np.sqrt((X-xv)**2+(Y-yv)**2))

    return psi
```

```python
def vel_vort_combined(strength, xv, yv, X, Y, N):
   

    u_all, v_all = np.zeros((N,N)), np.zeros((N,N))
    
    for i in range(0,len(strength)):
        u_v, v_v = vel_vortex(strength[i], xv[i], yv[i], X, Y)
        u_all += u_v
        v_all += v_v
    
    return u_all, v_all
```

```python
def sf_vort_combined(strength, xv, yv, X, Y, N):
    
    
    psi_v = np.zeros((N,N), dtype = float)
    psi_all = np.zeros((N,N), dtype = float)
    
    for i in range(0,len(strength)):
        psi_v = sf_vortex(strength[i], xv[i], yv[i], X, Y)
        psi_all += psi_v
    
    return psi_all
```

```python
#Finite row of vortices

u_vortices, v_vortices = vel_vort_combined(vort_strength, xv, yv, X, Y, N)
u = u_vortices
v = v_vortices

psi_vortices = sf_vort_combined(vort_strength, xv, yv, X, Y, N)
psi = psi_vortices
```

```python
x = np.linspace(xs,xe,N)
x_vortices = x[::N/(n-1)] #picking off x coordinates of vortex locations
y_vortices = np.zeros_like(x_vortices)
a = x_vortices[1] - x_vortices[0]
c = 2.0*math.pi/a

u_inf = ((gamma/(2.0*a)) * (np.sinh(c*Y)/(np.cosh(c*Y) - np.cos(c*X))))
v_inf = (-(gamma/(2.0*a)) * (np.sin(c*X)/(np.cosh(c*Y) - np.cos(c*X))))
```
