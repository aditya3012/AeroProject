# 1. Derivation for complex potential of a 2D doublet 

       Complex variable is generaally written as Z= x+iy

       from Cauchy's theorem we got
       
       ∂ϕ/∂x = ∂Ψ/∂y, ∂ϕ/∂y = -∂Ψ/∂x

       similarly w(z) = ϕ + i Ψ  -----------------> 1

          w(z) = W'(z)
          
               = ∂ϕ/∂x + i ∂Ψ/∂y
               
               = u - i v
 
         U = U∞
         V = 0
         
         Potential Velocity (ϕ) = U∞ x
         
       Uniform Stream Velocity (Ψ) = U∞ y
    
                           w(z) = U∞(x+i*y)
                                = U∞(z)
                               
       Now, considering the stream at an angle "α" to the x- axis
    
       we get,             U = U∞Cosα 
                            v = U∞sinα
                            
                       ϕ(x,y) = U∞Cosαx + U∞sinαy -------------->2
                       
                       Ψ(x,y) = U∞Cosαy - U∞sinαx -------------->3
                       
                       plugging 2 and 3 in equation 1, 
                       
                       we get, w(z) = U∞ (xCosα + ysinα) + i U∞(-xsinα + yCosα ) 
                               
                                    = U∞ (Cosα - i sinα)(x + iy)
                                    
                                    = U∞ z e^-iα,
                               
                       this shows the complex potential flow field for one direction.
                       when the same flowfield is rotated by α in counterclockwise
                       
                        then we replace z*e^-iα = Z
                        
                        
                          we get, w(z) = U∞ * Z 
                          
                consider the  source with a strength k, at origin.
                        using polar coordinates
                        
                         we get w(z) = (σ/2*π)*(lnr + i*θ)
                            
                                     = (σ/2*π)ln(z)
                                     
                        similarly the equation for sink becomes 
                                    
                                     = (-σ/2*π)ln(z)
                                     
               on applying the limits to source and sink of an equal strength we get doubet
               
                                w(z) = (σ / 2*π)ln(z+a) + (σ / 2*π)ln(z-a)
                                    
                                     = (σ / 2*π)ln(z+a/z-a)
                                     
                                     = (σ / 2*π)ln {(1+a/z)/(1-a/z)}
                                     
                  on expanding  ln {(1+a/z)/(1-a/z)}
                
                  we get 2a (1+ a^2/ 3z^2 + a^ 4/5z^4....)
                  
                                 now on plugging into the equation,
                                 
                                 w = (2a σ / 2*π *z)(1 + a^2/ 3z^2 + a^ 4/5z^4....)
                                 
                                 lim (2a σ / 2*π *z)(1 + a^2/ 3z^2 + a^ 4/5z^4....)
                              a->0 2aσ->k
                           
                                    => 2a σ = k
                                    
                                    => k / 2*π *z
                                    
                               thefore the complex potential becomes
                               
                                    =>  k / 2*π *z
                                     
                       
                       
                       
                 
                      

# 2. Flow around the corner, using Complex potential

                                W(Z) = Z ^ n

## a. if n is a positive integer, then the flow field is divided into 2n equal parts.


              Given W(Z) = Z ^ n
     
              w(z) = ϕ + i Ψ
           
              writing Z in polar coordinates,
      
              Z = x + i y
            
            = r * (e^i*θ)
            
              therefore,
              e^i*θ = rcos θ + i*r*sinθ
           
              W(z) = r^n cosnθ + i*r^n*sinnθ
           
              Ψ  = r^n*sin nθ
             
              on equating stream line = 0
             
              r^n*sin nθ = 0
             
              considering when n = 2
             
              we get 
             
           => r^2*sin 2θ
          
           => r^2*(2sinθcosθ)
          
           => r^2*(2*x/r*y/r)
          
           => Ψ = 2xy
             
              for n = 2 we have 4 flow fields.

## 3.b. Pressure gradient along the wall

      From the bernoullis equation we have pressure at wall
      
       [P / ρ + (U ^2 * v ^ 2)]---------------->1
       
       Velocit U is 0 near the wall
       
       therefore, the equation 1. becomes
       
       [P / ρ + ( v ^ 2)] --------------------->2
       
       pressure for the stream 
       
       [P / ρ + (U ^2 * v ^ 2)]ranging from o to ∞ -->3
       
       and the velocities U and V is 0,
       
       therefore, the equation becomes, 
       
       [P ∞/ ρ ] -------------------------------->4
       
       on equating 2 and 4, and on solving them we get,
       
       [P - P ∞ / ρ ] = ( v ^2 /2) -------------->5
       
       but v is the imaginary part
       
       => v = (-Γ / 2 π)*(y / y^2 + a^2)--------->6
       
       plugging 6 in 5
       
       we get,
       
       [P - P ∞ / ρ ] = (1/2) [ (-Γ / 2 π)*(y / y^2 + a^2)] ^ 2
       
       at Γ = 2 and a = 1
       
       we get, [P - P ∞ / ρ ] = [2y^2/ π * (y^2 + 1)^2]
